# Building a Decision Tree Model using Python to predict if a person have diabetes

Steps:
- Import packages needed 
- Import the dataset
- Perform feature selection
- Split the data 
- Train the Decision Tree model
- Study the accuracy, precision, and recall of the model
- Visualize the tree

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn import metrics 
import matplotlib.pyplot as plt
```

```python
file = pd.read_csv('diabetes.csv')
df=pd.DataFrame(file)
df.head()
```

```python
headers = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']

df.columns=headers

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
X = df.drop(['label'], axis=1) 
y = df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier()

clf = clf.fit(X_train,y_train)

y_pred = clf.predict(X_test)

result_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})

print(result_df)
```

```python
accuracy= metrics.accuracy_score(y_test,y_pred)
precision= metrics.precision_score(y_test,y_pred)
recall= metrics.recall_score(y_test,y_pred)

print("Accuracy:",accuracy,"\n Precision: ",precision,"\n Recall: ",recall)
```

```python
plt.figure(figsize=(150, 100))
plot_tree(clf, filled=True, feature_names=X_train.columns, class_names=True)
plt.show()
```
